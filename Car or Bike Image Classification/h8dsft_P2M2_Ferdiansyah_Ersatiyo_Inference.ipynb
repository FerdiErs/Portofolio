```python
# Load the model
model = load_model('model.h5')
```

```python
# load the img1
img1 = image.load_img('image1.jpg')
img1
```

```python
# load the img2
img2 = image.load_img('image2.jpg')
img2
```

```python
# load the img3
img3 = image.load_img('image3.jpg')
img3
```

```python
# Change the shape into 200x200
# Shape of img1
img1 = image.img_to_array(img1) / 255
img1 = tf.image.resize(img1, (200,200))
img1 = tf.expand_dims(img1, axis=0)


# Change the shape into 220x220
img2 = image.img_to_array(img2) / 255
img2 = tf.image.resize(img2, (200,200))
img2 = tf.expand_dims(img2, axis=0)

# Change the shape into 220x220
img3 = image.img_to_array(img3) / 255
img3 = tf.image.resize(img3, (200,200))
img3 = tf.expand_dims(img3, axis=0)
```

```python
prediction1 = model.predict(x = img1)
prediction2 = model.predict(x = img2)
prediction3 = model.predict(x = img3)

def Predict(value):
    if value >= 0.5:
        return "Car"
    else:
        return "Bike"


print('')
print('Image 1 Prediction : ', Predict(prediction1))
print('Image 2 Prediction : ', Predict(prediction2))
print('Image 3 Prediction : ', Predict(prediction3))
```
