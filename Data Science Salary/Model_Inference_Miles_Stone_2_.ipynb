# Model Prediction with a new data

```python
import pandas as pd
import joblib as jb
```

```python
inf = {
  'work_year': [2023, 2020],
  'experience_level': ['SE', 'MI'],
  'employment_type':['FT', 'FL'],
  'job_title' : ['Data Engineer','ML Engineer'],
  'employee_residence':['ID', 'JP'],
  'remote_ratio': [100, 100],
  'company_location': ['ID', 'JP'],
  'company_size': ['L','L']
}
df = pd.DataFrame(inf)

display(df)
```

```python
# Loading Model
model_inf = jb.load('/content/gradient_boosting_model.pkl')
```

```python
# predict data inference with model
data_inf_pred = model_inf.predict(df)
```

```python
# prediction result
print(data_inf_pred)
```

```python
# create prediction to dataframe
yPredInfDF = pd.DataFrame(data_inf_pred, columns=['Salary Prediction'])
yPredInfDF
```

```python
# Concate datainference and data predict
finalPredDF = pd.concat([df.reset_index(drop=True), yPredInfDF.reset_index(drop=True)], axis=1 )
finalPredDF
```
